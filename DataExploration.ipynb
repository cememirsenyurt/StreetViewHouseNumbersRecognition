```python
import numpy as np
import matplotlib.pyplot as plt

from helpers import loadData
```

```python
X_tr, y_tr = loadData("train")
print(type(X_tr))
print(type(y_tr))
```

```python
# Display 10 random training images and their label
n = X_tr.shape[3]
s = 10
f, ax = plt.subplots(s//2, 2, figsize=(32, 32))
rand_ind = np.random.randint(0, high=n, size=s)
for i, ind in zip(range(s),rand_ind):
    ax[i//2][i%2].imshow(X_tr[:,:,:,ind])
    ax[i//2][i%2].set_title("Label: {}".format(y_tr[ind][0]))
```
